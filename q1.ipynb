```python
# data for attribute avgs
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter

dataset = pd.read_csv("datasets/facebook_data.csv")
# strip empty columns
del dataset['Column1']
del dataset['Column2']
del dataset['Column3']
del dataset['Column4']

# sort data by earliest status publish data
dataset['status_published'] = pd.to_datetime(dataset['status_published'])
dataset = dataset.sort_values(by="status_published")
dataset.head()
```

### Scatter Plot for the Number of Reactions & the Number of Shares
##### *Figures 1 & 2*

```python
# scatter plot for num_reactions vs num_likes
plt.scatter(dataset.num_reactions, dataset.num_likes, color='#9fe2bf')
plt.title("Correlation of Number of Reactions vs Number of Likes")
plt.xlabel("Number of Reactions")
plt.ylabel("Number of Likes")

# plot line of best fit
plt.plot(np.unique(dataset.num_reactions), np.poly1d(np.polyfit(dataset.num_reactions, dataset.num_likes, 1))(np.unique(dataset.num_reactions)), color = 'black')
plt.show()

# scatter plot for num_shares vs num_likes
plt.scatter(dataset.num_shares, dataset.num_likes, color='#e29fc2')
plt.title("Correlation of Number of Shares vs Number of Likes")
plt.xlabel("Number of Shares")
plt.ylabel("Number of Likes")

# plot line of best fit
plt.plot(np.unique(dataset.num_shares), np.poly1d(np.polyfit(dataset.num_shares, dataset.num_likes, 1))(np.unique(dataset.num_shares)), color = 'black')
plt.show()
```

### Histogram for the Number of Reactions & the Number of Shares
##### *Figures 3 & 4*

```python
# histogram for num_reactions
plt.figure(figsize=(8, 6), dpi=80)
plt.hist(dataset.num_reactions, color='#e1b941', bins=75, density=True, alpha=0.7, rwidth=0.85)
plt.title("Distribution of Number of Reactions on a Post")
plt.xlabel("Num Reactions")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# histogram for num_shares
plt.figure(figsize=(8, 6), dpi=80)
plt.hist(dataset.num_shares, color='#4169e1', bins=30, density=True, alpha=0.7, rwidth=0.92)
plt.title("Distribution of Number of Shares on a Post")
plt.xlabel("Num Shares")
plt.ylabel("Frequency") # This is the number of posts
plt.grid()
plt.show()
```

### Boxplot for the Number of Reactions & the Number of Shares
##### *Figures 5 & 6*

```python
# boxplot for num_reactions
plt.figure(figsize=(6, 6))
plt.boxplot(dataset.num_reactions, showfliers=False)
plt.title("Boxplot of Number of Reactions on a Post")
plt.ylabel("Number of Reactions")
plt.show()

# boxplot for num_shares
plt.figure(figsize=(6, 6))
plt.boxplot(dataset.num_shares, showfliers=False)
plt.title("Boxplot of Number of Shares on a Post")
plt.ylabel("Number of Shares")
plt.show()
```
