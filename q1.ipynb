```python
# data for attribute avgs
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter

dataset = pd.read_csv("datasets/facebook_data.csv")
# strip empty columns
del dataset['Column1']
del dataset['Column2']
del dataset['Column3']
del dataset['Column4']

# sort data by earliest status publish data
dataset['status_published'] = pd.to_datetime(dataset['status_published'])
dataset = dataset.sort_values(by="status_published")

# show data
xcol1 = pd.DataFrame(dataset, columns = ["num_loves"])
xcol2 = pd.DataFrame(dataset, columns = ["num_likes"])
ycol = pd.DataFrame(dataset, columns = ["num_reactions"])
dataset.head()
```

```python
# scatter plot for num_reactions vs num_shares

plt.scatter(dataset.num_reactions, dataset.num_shares)
plt.title("Scatterplot of Number of Reactions vs Number of Shares")
plt.xlabel("Num Reactions")
plt.ylabel("Num Shares")
plt.show()


# scatter plot for num_likes vs num_shares
plt.scatter(dataset.num_likes, dataset.num_shares)
plt.title("Scatterplot of Number of Likes vs Number of Shares")
plt.xlabel("Num Likes")
plt.ylabel("Num Shares")
plt.show()

plt.scatter(xcol2, ycol)
plt.title("num_likes vs num_reactions")
plt.xlabel("num_likes")
plt.ylabel("num_reactions")
plt.xlim(-40 , 2000)
plt.ylim(-300 , 2500)
plt.show()

# note to self for report from jonathan:
# the graphs
```

```python
# histogram for num_reactions
plt.figure(figsize=(8, 6), dpi=80)
plt.hist(dataset.num_reactions, color='#FFC0CB', edgecolor='black')
plt.title("Distribution of Number of Reactions on a Post")
plt.xlabel("Num Reactions")
plt.ylabel("Frequency")
#plt.xlim(-10, 400)
# make x axis ticks every 25
#plt.xticks(np.arange(0, 400, 50))
plt.show()

plt.figure(figsize=(8, 6), dpi=80)
plt.hist(dataset.num_likes, color='#c0cbff', edgecolor='black')
plt.title("Distribution of Number of Likes on a Post")
plt.xlabel("Num Likes")
plt.ylabel("Frequency") # This is the number of posts
# plt.xlim(-10, 400)
# plt.ylim(0, 3000)
plt.show()
```

```python
# boxplot for num_reactions
plt.figure(figsize=(10, 10))
plt.boxplot(dataset.num_reactions, showfliers=False)
plt.title("Boxplot of Number of Reactions on a Post")
plt.ylabel("Num Reactions")
plt.show()

# boxplot for num_likes
plt.figure(figsize=(10, 10))
plt.boxplot(dataset.num_likes, showfliers=False)
plt.title("Boxplot of Number of Likes on a Post")
plt.ylabel("Num Likes")
plt.show()
```
